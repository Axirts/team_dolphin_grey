```python
hhs['Payment'] = hhs['Payment'].str.replace('[$,]','')
```

```python
hhs['Payment'] = hhs['Payment'].str.strip()
```

```python
hhs['Payment'] = hhs['Payment'].astype('float64')
```

```python
hhs.head(3)
```

```python
hhs.dtypes
```

```python
def convert_currency_to_float(dataframe,columns):
    for col in columns:
        dataframe[col] = dataframe[col].str.replace('[$,-]','')
        dataframe[col] = dataframe[col].str.strip()
        dataframe[col] = dataframe[col].astype('float64')
```

```python
convert_currency_to_float(claims,["Claims Paid for Testing","Claims Paid for Treatment","Claims Paid for Vaccine"])
```
